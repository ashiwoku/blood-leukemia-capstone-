```python
import numpy as np
import os
import random
import cv2
import sys
import matplotlib.pyplot as plt
```

```python
## Load in image folder.
data_directory = 'C:\\Users\\User\\Desktop\\leukemia capstone'
categories = ['BAS','EBO','EOS','KSC','LYA','LYT','MMZ','MOB','MON',
              'MYB','NGS','MYO','NGB','PMB','PMO']
```

```python
img_folder = []
class_folder = []
def create_dataset(dataset):
    for label in dataset:
        path = os.path.join(data_directory,label)
        ##classification = categories.index(label)
        for image in random.sample(os.listdir(path),8):
            try:
                image_array = cv2.imread(os.path.join(path,image),0)
                image_array = image_array/255
                img_folder.append(image_array.flatten())
                class_folder.append(label)
            except Exception as e:
                print("Image was a problem {}".format(image))
```

```python
create_dataset(categories)
```

```python
BAS = img_folder[0:8]
EBO =img_folder[8:16]
EOS =img_folder[16:24]
KSC =img_folder[24:32]
LYA =img_folder[32:40]
LYT =img_folder[40:48]
MMZ =img_folder[48:56]
MOB =img_folder[56:64]
MON =img_folder[64:72]
MYB =img_folder[72:80]
NGS =img_folder[80:88]
MYO =img_folder[88:96]
NGB =img_folder[96:104]
PMB =img_folder[104:112]
PMO =img_folder[112:120]
```

```python
def print_images(image_set):
    columns = 3
    row =3
    for i in range(1,columns*row -1):
        plt.subplot(columns,row,i)
        img = np.reshape(image_set[i-1],(400,400))
        plt.imshow(img)
    plt.show()
```
